```python
import cv2
import os

video_folder = "/content/videos"
output_folder = "/content/dataset"
os.makedirs(output_folder, exist_ok=True)

# Extract frames every 0.25 seconds (250ms)
interval_ms = 250

for video in os.listdir(video_folder):
    if video.endswith(".mp4"):
        video_path = os.path.join(video_folder, video)

        # student id = filename without extension
        student_id = os.path.splitext(video)[0]
        student_folder = os.path.join(output_folder, student_id)
        os.makedirs(student_folder, exist_ok=True)

        cap = cv2.VideoCapture(video_path)
        frame_count = 0

        while True:
            cap.set(cv2.CAP_PROP_POS_MSEC, frame_count * interval_ms)
            ret, frame = cap.read()

            if not ret:
                break

            save_path = os.path.join(student_folder, f"{frame_count}.jpg")
            cv2.imwrite(save_path, frame)
            frame_count += 1

        cap.release()
        print(f"‚úîÔ∏è Extracted {frame_count} frames for {student_id}")

print("\nüéâ All frames extracted successfully!")
```
